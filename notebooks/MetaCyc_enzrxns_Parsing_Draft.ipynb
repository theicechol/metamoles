```python
import pandas as pd
import numpy as np
```

```python
# I hack a .dat file by opening in excel and save it as .csv file which I'm more confident in
csv = pd.read_csv('enzrxns.csv', skiprows = 59, names=['Col'])
#csv.columns('Col')
```

```python
csv.head(50)
```

```python
k1 = 0
ID = [] # 'UNIQUE-ID - '
ID_L = len('UNIQUE-ID - ')

k2 = 0
CName = [] # COMMON-NAME - 
CName_L = len('COMMON-NAME - ')

k3 = 0
enz = [] # ENZYME - 
enz_L = len('ENZYME - ')

k4 = 0
rxn = [] # REACTION - 
rxn_L = len('REACTION - ')

k5 = 0
alt = [] # ALTERNATIVE-SUBSTRATES - 
alt_L = len('ALTERNATIVE-SUBSTRATES - ')

substrate = []
reaction = []
for i in range(csv.shape[0]):
    
    data = csv['Col'][i]
    if data[:ID_L] == 'UNIQUE-ID - ':
        ID.append(data[(ID_L):])
        k1 += 1
        #print(data[(ID_L):])
    elif data[:CName_L] == 'COMMON-NAME - ':
        CName.append(data[(CName_L):])
        k2 += 1
    elif data[:enz_L] == 'ENZYME - ':
        enz.append(data[(enz_L):])
        k3 += 1
    elif data[:rxn_L] == 'REACTION - ':
        reaction.append(data[(rxn_L):])
        # k4 += 1
    elif data[:alt_L] == 'ALTERNATIVE-SUBSTRATES - ':
        substrate.append(data[(alt_L):]) # there are several substrate which will be appended to another list first
        # k5 += 1
    else:
        pass
    
    if k2 < k1-1:
        CName.append('N/A')
        k2 += 1
    if k3 < k1-1:
        enz.append('N/A')
        k3 += 1
    if k4 < k1-1:
        if reaction != []:
            rxn.append(reaction)
            reaction = []
        else:
            rxn.append('N/A')
        k4 += 1
    if k5 < k1-1:
        #print(substrate)
        if substrate != []:
            alt.append(substrate)
            substrate = []
        else:
            alt.append('N/A')
        k5 += 1

# double check row count

if k4 < k1:
    if reaction != []:
        rxn.append(reaction)
        reaction = []
    else:
        rxn.append('N/A')
    k4 += 1
if k5 < k1:
    #print(substrate)
    if substrate != []:
        alt.append(substrate)
        substrate = []
    else:
        alt.append('N/A')
    k5 += 1

print ('Number of compounds count is', k1, k2, k3, 'rxn', k4, 'and alt-subs', k5)
```

```python
df_enzrxns = pd.DataFrame({'UNIQUE-ID' : ID,
                    'COMMON-NAME' : CName,
                    'ENZYME' : enz,
                    'REACTION' : rxn,
                    'ALTERNATIVE-SUBSTRATES' : alt})
```

```python
df_enzrxns.to_csv('df_enzrxns.csv')
```
