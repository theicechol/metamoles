```python
k1 = 0
ID = [] # 'UNIQUE-ID - '
ID_L = len('UNIQUE-ID - ')

k2 = 0
CName = [] # COMMON-NAME - 
CName_L = len('COMMON-NAME - ')

k3 = 0
Gibb = [] # GIBBS-0 - 
Gibb_L = len('GIBBS-0 - ')

k4 = 0
InChI = [] # INCHI - or NON-STANDARD-INCHI - 
InChI_L = len('INCHI - ')
NonInChI_L = len('NON-STANDARD-INCHI - ')

k5 = 0
SMILES = [] # SMILES - 
SMILES_L = len('SMILES - ')

for i in range(csv.shape[0]):
    
    data = csv['Col'][i]
    if data[:ID_L] == 'UNIQUE-ID - ':
        ID.append(data[(ID_L):])
        k1 += 1
    elif data[:CName_L] == 'COMMON-NAME - ':
        CName.append(data[(CName_L):])
        k2 += 1
    elif data[:Gibb_L] == 'GIBBS-0 - ':
        Gibb.append(data[(Gibb_L):])
        k3 += 1
    elif data[:InChI_L] == 'INCHI - ':
        InChI.append(data[(InChI_L):])
        k4 += 1
    elif data[:SMILES_L] == 'SMILES - ':
        SMILES.append(data[(SMILES_L):])
        k5 += 1
    
    # in case that there is no InChI available
    elif data[:NonInChI_L] == 'NON-STANDARD-INCHI - ' and k4 < k1:
        InChI.append(data[(NonInChI_L):])
        k4 += 1
    else:
        pass
    
    if k2 < k1-1:
        CName.append('N/A')
        k2 += 1
    if k3 < k1-1:
        Gibb.append('N/A')
        k3 += 1
    if k4 < k1-1:
        InChI.append('N/A')
        k4 += 1
    if k5 < k1-1:
        SMILES.append('N/A')
        k5 += 1

print ('Number of compounds count is', k1, k2, k3, k4, 'and', k5)
```

```python
df_cpd = pd.DataFrame({'UNIQUE-ID' : ID,
                       'COMMON-NAME' : CName,
                       'GIBBS-0' : Gibb,
                       'INCHI' : InChI,
                       'SMILES' : SMILES})
```

```python
df_cpd.to_csv('df_cpd.csv')
df_cpd
```
